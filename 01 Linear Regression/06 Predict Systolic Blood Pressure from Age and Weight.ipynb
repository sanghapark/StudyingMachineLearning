```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def fit(X, Y):
    w = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
    YHat = X.dot(w)
    ss_rss = np.dot(Y - YHat, Y - YHat)
    ss_tss = np.dot(Y - Y.mean(), Y - Y.mean())
    return w, YHat, 1 - (ss_rss/ss_tss)
```

```python
df = pd.read_excel('mlr02.xls')

df['ones'] = 1

w_2, YHat_2, r2_2 = fit(df[['X2', 'ones']].as_matrix(), df['X1'].as_matrix())
w_3, YHat_3, r2_3 = fit(df[['X3', 'ones']].as_matrix(), df['X1'].as_matrix())
w_23, YHat_23, r2_23 = fit(df[['X2', 'X3', 'ones']].as_matrix(), df['X1'].as_matrix())


plt.scatter(df['X2'], df['X1'])
plt.plot(df['X2'], df[['X2', 'ones']].as_matrix().dot(w_2), c='r')
plt.show()
plt.scatter(df['X3'], df['X1'])
plt.plot(df['X3'], df[['X3', 'ones']].as_matrix().dot(w_3), c='r')
plt.show()


print(r2_2, r2_3, r2_23)

```
