# L1 vs L2 Regularization

* L1: Encourages a sparse solution
  * Few w's non-zero, many equal to 0
  * Choose the most important features
* L2: Encourages small weights
  * All w's close to 0, but not exactly 0
  * Make the assertion that none of the weights are extremely large

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

xx = np.linspace(-10, 10, 1000)
plt.plot(xx, xx**2, label='$w^2$')
plt.plot(xx, np.abs(xx), label='|w|')
plt.ylim((-1, 3))
plt.xlim((-3, 3))
plt.legend()
```

* L2 penalty is quadratic, and L1 is an absolute function.
* Key is the derivative.
* L2: as w goes to 0, derivatieve goes to 0
  * If w is small, further gradient descent won't change it much.
* L1: Derivative is always 1 or -1 (0 at w=0)
  * No matter where w is. It will fall at a constant rate.
  * When it reaches 0, it will stays there forever.
