```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
for line in open('data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x**2])
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)


w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))

YHat = X.dot(w)

plt.scatter(X[:, 1], Y)
plt.scatter(X[:, 1], YHat, c='r')

ss_rss = np.dot(Y - YHat, Y - YHat)
ss_tss = np.dot(Y - Y.mean(), Y - Y.mean())
r2 = 1 - (ss_rss/ss_tss)
r2
```
