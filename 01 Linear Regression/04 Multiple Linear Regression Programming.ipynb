```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

# load the data
X = []
Y = []
for line in open('data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([1, float(x1), float(x2)])
    Y.append(float(y))
X = np.array(X)
Y = np.array(Y)


# plot
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.scatter(X[:,0], X[:,1], Y)
# plt.show()

# calculate weights
w = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
Yhat = X.dot(w)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
ax.scatter(X[:,0], X[:,1], Yhat, c='r')
plt.show()

ss_res = (Y - Yhat).dot(Y - Yhat)
ss_tot = (Y - Y.mean()).dot(Y - Y.mean())
r2 = 1 - (ss_res/ss_tot)
r2
```
